```python
import cv2
import matplotlib.pyplot as plt
import easyocr
```

```python
img = cv2.imread('inputs\image2.jpg')
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
reader_fa = easyocr.Reader(['fa']) # this needs to run only once to load the model into memory
result = reader_fa.readtext('inputs\image2.jpg')
result
```
